# Задание 2

```python
lower_bound = Q1-(1.5*IQR)
lower_bound
```

```python
upper_bound = Q3+(1.5*IQR) 
upper_bound
```

# Выбросы по температуре 

```python
remove_outliers_temp = df_data_new1[df_data_new1['Температура'].between(lower_bound['Температура'], 
                  upper_bound['Температура'], inclusive = True)].sort_values('Температура')
remove_outliers_temp #выбросов по Температуре 74 
```

```python
outliers_temp = pd.concat([df_data_new1,remove_outliers_temp]).drop_duplicates(keep= False)
outliers_temp #получили выбросы по температуре (Нормальной температурой для лошадей
#является 37.5-38.5, данные выбросы не являются нормой, но для принятия решения стоит оставить данную информацию,
#она говорит о заболеваниях)
```

# Выбросы по пульсу

```python
remove_outliers_pulse = df_data_new1[df_data_new1['Пульс'].between(lower_bound['Пульс'], 
                  upper_bound['Пульс'], inclusive = True)].sort_values('Пульс')
remove_outliers_pulse #выбросов по Пульсу 29 
```

```python
outliers_pulse = pd.concat([df_data_new1,remove_outliers_pulse]).drop_duplicates(keep= False)
outliers_pulse #получили выбросы по пульсу (Нормальный пульс у лошадей 25-44 удара в минуту,
#данная информация хранит пустые значения и не часто встречающиеся данные, можем не использовать их далее)
```

# Выбросы по дыханию 

```python
remove_outliers_respiratory_rate = df_data_new1[df_data_new1['Дыхание'].between(lower_bound['Дыхание'], 
                  upper_bound['Дыхание'], inclusive = True)].sort_values('Дыхание')
remove_outliers_respiratory_rate #выбросов по Дыханию 75 
```

```python
outliers_pulse_respiratory_rate = pd.concat([df_data_new1,remove_outliers_respiratory_rate]).drop_duplicates(keep= False)
outliers_pulse_respiratory_rate #получили выбросы по дыханию (Нормальная частота дыхания у лошадей 8-12 в минуту,
#данная информация хранит пустые значения и не совсем реальные данные, можем не использовать их далее)
```
